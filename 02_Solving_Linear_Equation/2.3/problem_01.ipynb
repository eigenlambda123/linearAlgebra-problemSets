### **Problem**

Write down the 3×3 matrices that produce these elimination steps:

- **(a)** $E_{21}$: Subtract 5 times row 1 from row 2
- **(b)** $E_{32}$: Subtract −7 times row 2 from row 3
- **(c)** $P$: Exchange rows 1 and 2, then rows 2 and 3

---

### **Solution**

**(a)** $E_{21}$: Row operation — $R_2 \leftarrow R_2 - 5R_1$

$$
E_{21} = 
\begin{bmatrix}
1 & 0 & 0 \\
-5 & 1 & 0 \\
0 & 0 & 1
\end{bmatrix}
$$


**(b)** $E_{32}$: Row operation — $R_3 \leftarrow R_3 + 7R_2$ (since subtracting −7 is adding 7)

$$
E_{32} = 
\begin{bmatrix}
1 & 0 & 0 \\
0 & 1 & 0 \\
0 & 7 & 1
\end{bmatrix}
$$


**(c)** $P$: First swap rows 1 and 2, then swap rows 2 and 3

This can be represented as a product of two permutation matrices:

* Swap rows 1 and 2:

  $$
  P_1 = 
  \begin{bmatrix}
  0 & 1 & 0 \\
  1 & 0 & 0 \\
  0 & 0 & 1
  \end{bmatrix}
  $$

* Swap rows 2 and 3:

  $$
  P_2 = 
  \begin{bmatrix}
  1 & 0 & 0 \\
  0 & 0 & 1 \\
  0 & 1 & 0
  \end{bmatrix}
  $$

Final permutation matrix:

$$
P = P_2 \cdot P_1 = 
\begin{bmatrix}
0 & 1 & 0 \\
1 & 0 & 0 \\
0 & 0 & 1
\end{bmatrix}
$$

(Note: The final result swaps rows 1 and 2, and row 3 stays in place due to the swap reversals.
