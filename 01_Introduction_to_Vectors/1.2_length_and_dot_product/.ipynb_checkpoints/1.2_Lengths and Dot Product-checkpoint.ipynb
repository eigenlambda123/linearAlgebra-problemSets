### <span style="color:red">Problem 1.2 (Calculating Dot Products)</span>

**Problem Statement 1:**

![Problem 1.2 – Calculate Dot Products](images/1.2%20Calculate%20Dot%20Products.png)


**Initial Approach**:

The problem asks for four dot product calculations involving vectors **u**, **v**, and **w**. I computed each as follows:

1. **u · v**
   $$
   \mathbf{u} \cdot \mathbf{v} = (-0.6)(4) + (0.8)(3) = -2.4 + 2.4 = \boxed{1.8}
   $$

2. **u · w**
   $$
   \mathbf{u} \cdot \mathbf{w} = (-0.6)(1) + (0.8)(2) = -0.6 + 1.6 = \boxed{1}
   $$

3. **u · (v + w)**
   - First compute:  
     $$
     \mathbf{v} + \mathbf{w} = (4 + 1, 3 + 2) = (5, 5)
     $$
   - Then:  
     $$
     \mathbf{u} \cdot (\mathbf{v} + \mathbf{w}) = (-0.6)(5) + (0.8)(5) = -3 + 4 = \boxed{1}
     $$

4. **v · w**
   $$
   \mathbf{v} \cdot \mathbf{w} = (4)(1) + (3)(2) = 4 + 6 = \boxed{10}
   $$

**Final Answers**:

- **u · v** = **1.8**  
- **u · w** = **1**  
- **u · (v + w)** = **1**  
- **v · w** = **10**


**Topic/Idea:**

- Dot Product Definition and Computation  
- Distributive Property of Dot Product:  
  $\mathbf{u} \cdot (\mathbf{v} + \mathbf{w}) = \mathbf{u} \cdot \mathbf{v} + \mathbf{u} \cdot \mathbf{w}$

### <span style="color:red">Problem 1.2 (*Get Lengths and Check Schwarz Inequalities*)</span>

**Problem Statement 2:**

Use the same vectors from Problem 1.2 (Calculating Dot Products) to compute their lengths and verify the Schwarz inequality:

![Problem 1.2 – Get Lengths](images/1.2%20Get%20Lengths%20Problem.png)


**Initial Approach:**

* The length (or norm) of a vector **v** is computed as:

  $$
  \|\mathbf{v}\| = \sqrt{\mathbf{v} \cdot \mathbf{v}}
  $$
* So it's easier to first compute the square of the length (**‖v‖²**), then take the square root if needed.
* Once lengths are known, I checked the **Schwarz inequality**:

  $$
  |\mathbf{a} \cdot \mathbf{b}| \leq \|\mathbf{a}\| \cdot \|\mathbf{b}\|
  $$

  for two combinations: **u · v** and **v · w**.


**Final Answer:**

* **Vector Lengths:**

  1. ‖u‖²

     $$
     (-0.6)^2 + (0.8)^2 = 0.36 + 0.64 = 1 \Rightarrow \|\mathbf{u}\| = \sqrt{1} = 1
     $$

  2. ‖v‖²

     $$
     (4)^2 + (3)^2 = 16 + 9 = 25 \Rightarrow \|\mathbf{v}\| = \sqrt{25} = 5
     $$

  3. ‖w‖²

     $$
     (1)^2 + (2)^2 = 1 + 4 = 5 \Rightarrow \|\mathbf{w}\| = \sqrt{5} \approx 2.236
     $$

* **Schwarz Inequality Checks:**

  1. **|u · v| ≤ ‖u‖ · ‖v‖**

     $$
     |(-0.6)(4) + (0.8)(3)| = |-2.4 + 2.4| = |1.8| = 1.8 \leq 1 \cdot 5 = 5 \quad \text{✓ True}
     $$

  2. **|v · w| ≤ ‖v‖ · ‖w‖**

     $$
     |(4)(1) + (3)(2)| = |4 + 6| = 10 \leq 5 \cdot \sqrt{5} \approx 11.18 \quad \text{✓ True}
     $$


**Topic/Idea:**

* Vector Magnitude (Norm)
* Length Calculation
* Schwarz (Cauchy–Schwarz) Inequality
* Relationship Between Algebraic and Geometric Properties of Vectors

### <span style="color:red">Problem 1.2 – Find Unit Vectors and Angles</span>

**Problem Statement 3:**

Use vectors from Problem 1.2 (Calculating Dot Products) to:

1. Find the unit vectors of **v = (4, 3)** and **w = (1, 2)**
2. Compute the cosine and angle θ between them
3. Find vectors **a**, **b**, and **c** that form angles of 0°, 90°, and 180° with **w**

![Problem 1.2 –  Find Unit Vectors and Angles](images/1.2%20Find%20Unit%20Vectors.png)


### Step 1: Finding Unit Vectors

The formula for a unit vector:
  **v̂ = v / ||v||**

* For **v = (4, 3)**:
    ||v|| = √(4² + 3²) = √25 = 5
    → **v̂ = (4/5, 3/5)**

* For **w = (1, 2)**:
    ||w|| = √(1² + 2²) = √5
    → **ŵ = (1/√5, 2/√5)**


### Step 2: Finding Cosine and Angle Between v and w

The formula:
  **cos(θ) = (v · w) / (||v|| · ||w||)**

* Dot product:
    v · w = (4)(1) + (3)(2) = 10

* Magnitudes:
    ||v|| = 5  ||w|| = √5

* Cosine and angle:
    cos(θ) = 10 / (5√5) = **2/√5 ≈ 0.894**
    θ = cos⁻¹(0.894) ≈ **26.57°**


### Step 3: Finding Vectors That Form Specific Angles with w = (1, 2)

#### Angle = 0° (Same Direction)

* Use a positive scalar multiple of **w**:
    **a = k · w**, k > 0
    Choose k = 1/√5 → **a = (1/√5, 2/√5)**

* Verification:
    a · w = √5
    ||a|| = 1, ||w|| = √5
    cos(θ) = √5 / (1 · √5) = 1 → θ = 0°

#### Angle = 90° (Perpendicular)

* Condition: **a · w = 0**
    For a = (x, y), this gives: x + 2y = 0 → **x = –2y**

* General form:
    **a = (–2y, y)**
    Examples: (–2, 1) or (2, –1)

#### Angle = 180° (Opposite Direction)

* Use a negative scalar multiple of **w**:
    **a = –k · w**, k > 0
    Choose k = 1/√5 → **a = (–1/√5, –2/√5)**

* Verification:
    a · w = –√5
    ||a|| = 1
    cos(θ) = –1 → θ = 180°


### Final Answers:

* Unit Vectors:
    v̂ = (4/5, 3/5)
    ŵ = (1/√5, 2/√5)

* Cosine and Angle Between v and w:
    cos(θ) = 2/√5 ≈ 0.894
    θ ≈ 26.57°

* Vectors Forming Angles with w = (1, 2):
    • 0° → (1/√5, 2/√5)
    • 90° → any vector of the form (–2y, y), e.g., (–2, 1)
    • 180° → (–1/√5, –2/√5)


**Topic/Idea:**

* Unit Vector
* Cosine and Angle
* Relationship Between Algebraic and Geometric Properties of Vectors
