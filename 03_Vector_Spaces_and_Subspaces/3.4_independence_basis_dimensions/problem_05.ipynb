### **Problems**

**Decide the dependence or independence of:**

#### **(a)**
Vectors:
$$ \mathbf{v}_1 = \begin{pmatrix}1 \\ 3 \\ 2\end{pmatrix},\quad \mathbf{v}_2 = \begin{pmatrix}2 \\ 1 \\ 3\end{pmatrix},\quad \mathbf{v}_3 = \begin{pmatrix}3 \\ 2 \\ 1\end{pmatrix} $$

#### **(b)**
Vectors:
$$ \mathbf{w}_1 = \begin{pmatrix}1 \\ -3 \\ 2\end{pmatrix},\quad \mathbf{w}_2 = \begin{pmatrix}2 \\ 1 \\ -3\end{pmatrix},\quad \mathbf{w}_3 = \begin{pmatrix}-3 \\ 2 \\ 1\end{pmatrix} $$


---

### **Solutions**

#### **Vectors $(1,3,2)$, $(2,1,3)$, $(3,2,1)$**

**Construct matrix $A$ with vectors as columns:**
$$
A = \begin{bmatrix}
1 & 2 & 3 \\
3 & 1 & 2 \\
2 & 3 & 1
\end{bmatrix}
$$

**Compute determinant:**
$$
\begin{align*}
\det(A) &= 1(1 \cdot 1 - 2 \cdot 3) - 2(3 \cdot 1 - 2 \cdot 2) + 3(3 \cdot 3 - 1 \cdot 2) \\
&= 1(1-6) - 2(3-4) + 3(9-2) \\
&= -5 + 2 + 21 \\
&= 18 \neq 0
\end{align*}
$$

**Conclusion:** $\boxed{\text{Linearly independent}}$ (since $\det(A) \neq 0$)

#### **Vectors $(1,-3,2)$, $(2,1,-3)$, $(-3,2,1)$**

**Construct matrix $B$:**
$$
B = \begin{bmatrix}
1 & 2 & -3 \\
-3 & 1 & 2 \\
2 & -3 & 1
\end{bmatrix}
$$

**Compute determinant:**
$$
\begin{align*}
\det(B) &= 1(1 \cdot 1 - 2 \cdot (-3)) - 2(-3 \cdot 1 - 2 \cdot 2) + (-3)(-3 \cdot (-3) - 1 \cdot 2) \\
&= 1(1+6) - 2(-3-4) - 3(9-2) \\
&= 7 + 14 - 21 \\
&= 0
\end{align*}
$$

**Verification via Row Reduction:**
$$
\begin{bmatrix}
1 & 2 & -3 \\
0 & 7 & -7 \\
0 & -7 & 7
\end{bmatrix}
\to
\begin{bmatrix}
1 & 2 & -3 \\
0 & 1 & -1 \\
0 & 0 & 0
\end{bmatrix}
$$
(Rank = 2 < 3)

**Conclusion:** $\boxed{\text{Linearly dependent}}$ (since $\det(B) = 0$)
