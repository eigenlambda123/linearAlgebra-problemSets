### Problems
Given a lower triangular matrix $A = L$:
$$
L = \begin{bmatrix} 
1 & 0 & 0 \\ 
a & 1 & 0 \\ 
b & c & 1 
\end{bmatrix}
$$
with elimination matrices:
- $E_{21}$ (eliminates $a$ with multiplier $-a$)
- $E_{31}$ (eliminates $b$ with multiplier $-b$)
- $E_{32}$ (eliminates $c$ with multiplier $-c$)

**Tasks:**

**(a)** Multiply $E_{32}E_{31}E_{21}$ to find the single matrix $E$ that reduces $A$ to $I$.

**(b)** Multiply $E_{21}^{-1}E_{31}^{-1}E_{32}^{-1}$ to reconstruct $L$.

---

### Solutions

#### **(a) Find $E = E_{32}E_{31}E_{21}$**
1. **Elimination Matrices:**
   $$
   E_{21} = \begin{bmatrix} 
   1 & 0 & 0 \\ 
   -a & 1 & 0 \\ 
   0 & 0 & 1 
   \end{bmatrix}, \quad
   E_{31} = \begin{bmatrix} 
   1 & 0 & 0 \\ 
   0 & 1 & 0 \\ 
   -b & 0 & 1 
   \end{bmatrix}, \quad
   E_{32} = \begin{bmatrix} 
   1 & 0 & 0 \\ 
   0 & 1 & 0 \\ 
   0 & -c & 1 
   \end{bmatrix}
   $$

2. **Multiply $E_{32}E_{31}E_{21}$:**
   - First, compute $E_{31}E_{21}$:
     $$
     E_{31}E_{21} = \begin{bmatrix} 
     1 & 0 & 0 \\ 
     -a & 1 & 0 \\ 
     -b & 0 & 1 
     \end{bmatrix}
     $$
   - Then multiply by $E_{32}$:
     $$
     E = E_{32}(E_{31}E_{21}) = \boxed{\begin{bmatrix} 
     1 & 0 & 0 \\ 
     -a & 1 & 0 \\ 
     -b & -c & 1 
     \end{bmatrix}}
     $$

3. **Verification:**
   - $EA = \begin{bmatrix} 
     1 & 0 & 0 \\ 
     -a & 1 & 0 \\ 
     -b & -c & 1 
     \end{bmatrix}
     \begin{bmatrix} 
     1 & 0 & 0 \\ 
     a & 1 & 0 \\ 
     b & c & 1 
     \end{bmatrix} = I \quad \checkmark$

#### **(b) Reconstruct $L = E_{21}^{-1}E_{31}^{-1}E_{32}^{-1}$**
1. **Inverse Elimination Matrices:**
   - $E_{21}^{-1} = \begin{bmatrix} 
     1 & 0 & 0 \\ 
     a & 1 & 0 \\ 
     0 & 0 & 1 
     \end{bmatrix}$
   - $E_{31}^{-1} = \begin{bmatrix} 
     1 & 0 & 0 \\ 
     0 & 1 & 0 \\ 
     b & 0 & 1 
     \end{bmatrix}$
   - $E_{32}^{-1} = \begin{bmatrix} 
     1 & 0 & 0 \\ 
     0 & 1 & 0 \\ 
     0 & c & 1 
     \end{bmatrix}$

2. **Multiply $E_{21}^{-1}E_{31}^{-1}E_{32}^{-1}$:**
   - First, compute $E_{31}^{-1}E_{32}^{-1}$:
     $$
     E_{31}^{-1}E_{32}^{-1} = \begin{bmatrix} 
     1 & 0 & 0 \\ 
     0 & 1 & 0 \\ 
     b & c & 1 
     \end{bmatrix}
     $$
   - Then multiply by $E_{21}^{-1}$:
     $$
     L = E_{21}^{-1}(E_{31}^{-1}E_{32}^{-1}) = \boxed{\begin{bmatrix} 
     1 & 0 & 0 \\ 
     a & 1 & 0 \\ 
     b & c & 1 
     \end{bmatrix}}
     $$
