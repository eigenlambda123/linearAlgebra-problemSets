### <span style="color:red">Problem 1.3 (*Find Linear Combination and Compute Dot Products*)</span>

**Problem Statement:**

![Problem 1.3 – Find Linear Combination and Compute Dot Products](images/1.3%20Find%20Linear%20Combination%20and%20Compute%20Dot%20Products.png)


**Initial Approach:**

* **Understanding the Problem:** There are three tasks:

  1. Find a **linear combination** that equals vector **b**
  2. Write **b** as a **matrix-vector multiplication** $Sx$
  3. Compute the **dot products** of the rows of matrix **S** with vector **x**


* Step 1: Express $b$ as a linear combination of vectors $s_1, s_2, s_3$:

  $$
  3s_1 + 4s_2 + 5s_3 = b
  $$

  Given:

  $$
  s_1 = (1,1,1), \quad s_2 = (0,1,1), \quad s_3 = (0,0,1)
  $$

  So:

  $$
  3(1,1,1) + 4(0,1,1) + 5(0,0,1) = (3, 7, 12)
  $$

* Step 2: Write $b$ as matrix-vector multiplication $Sx$:

  $$
  S = \begin{pmatrix}
  1 & 0 & 0 \\
  1 & 1 & 0 \\
  1 & 1 & 1 \\
  \end{pmatrix}, \quad x = \begin{pmatrix} 3 \\ 4 \\ 5 \end{pmatrix}
  $$

  Multiplying:

  $$
  Sx = \begin{pmatrix}
  1 & 0 & 0 \\
  1 & 1 & 0 \\
  1 & 1 & 1 \\
  \end{pmatrix} \begin{pmatrix} 3 \\ 4 \\ 5 \end{pmatrix} = \begin{pmatrix} 3 \\ 7 \\ 12 \end{pmatrix} = b
  $$

* Step 3: Compute dot products of each row of $S$ with $x$:

  $$
  \begin{aligned}
  \text{Row 1} \cdot x &= (1,0,0) \cdot (3,4,5) = 3 \\
  \text{Row 2} \cdot x &= (1,1,0) \cdot (3,4,5) = 3 + 4 = 7 \\
  \text{Row 3} \cdot x &= (1,1,1) \cdot (3,4,5) = 3 + 4 + 5 = 12 \\
  \end{aligned}
  $$


**Final Answer:**

* The linear combination:

  $$
  3s_1 + 4s_2 + 5s_3 = (3, 7, 12) = b
  $$

* Matrix-vector multiplication form:

  $$
  b = Sx, \quad \text{where} \quad S = \begin{pmatrix} 1 & 0 & 0 \\ 1 & 1 & 0 \\ 1 & 1 & 1 \end{pmatrix}, \quad x = \begin{pmatrix} 3 \\ 4 \\ 5 \end{pmatrix}
  $$

* Dot products of rows of $S$ with $x$:

  $$
  3, \quad 7, \quad 12
  $$


**Topic/Idea:**

* Linear Combination of Vectors
* Matrix-Vector Multiplication
* Dot Product Computation

### <span style="color:red">Problem 1.3 (*Find Unknown y*)</span>

**Problem Statement:**

![Problem 1.3 – Find Unknown y](images/1.3%20Find%20Unknown%20y.png)


**Initial Approach:**

* **Understanding the Problem:** Two tasks:

  1. Find the unknown vector $y$ such that $Sy = \mathbf{b}$ for given $\mathbf{b}$ vectors
  2. Answer the follow-up question about the sum of the first 5 odd numbers

* Task 1: Solve $Sy = (1,1,1)$

  The system of equations from matrix $S$:

  $$
  \begin{cases}
  y_1 = 1 \\
  y_1 + y_2 = 1 \implies y_2 = 0 \\
  y_1 + y_2 + y_3 = 1 \implies y_3 = 0 \\
  \end{cases}
  $$

  Solution:

  $$
  y = (1, 0, 0)
  $$

* Task 2: Solve $Sy = (1,4,9)$

  The system:

  $$
  \begin{cases}
  y_1 = 1 \\
  y_1 + y_2 = 4 \implies y_2 = 3 \\
  y_1 + y_2 + y_3 = 9 \implies y_3 = 5 \\
  \end{cases}
  $$

  Solution:

  $$
  y = (1, 3, 5)
  $$

* Task 3: Sum of the first 5 odd numbers:

  $$
  1 + 3 + 5 + 7 + 9 = 25
  $$

  (Alternatively, use the formula for sum of first $n$ odd numbers: $n^2$, where $n=5$, so sum = 25)


**Final Answer:**

* Solutions for $y$:

  $$
  y = (1, 0, 0) \quad \text{for} \quad Sy = (1,1,1)
  $$

  $$
  y = (1, 3, 5) \quad \text{for} \quad Sy = (1,4,9)
  $$

* Sum of first 5 odd numbers:

  $$
  25
  $$


**Topic/Idea:**

* Solving linear systems $Sy = b$
* Back-substitution method
* Sum of odd numbers and their formula $n^2$

### <span style="color:red">Problem 1.3 (*Find and Solve Equation*)</span>

**Problem Statement:**

![Problem 1.3 – Find and Solve Equation](images/1.3%20Find%20and%20Solve%20Equation.png)


**Initial Approach:**

* **Understanding the Problem:** Two tasks:

  1. Find $y$ in terms of $c$ (the unknown vector)
  2. Answer the question about column independence of matrix $S$

* Task 1: Solve the system $Sy = c$ given by equations:

  $$
  \begin{cases}
  y_1 = c_1 \\
  y_1 + y_2 = c_2 \\
  y_1 + y_2 + y_3 = c_3 \\
  \end{cases}
  $$

  From these:

  $$
  \begin{aligned}
  y_1 &= c_1 \\
  y_2 &= c_2 - y_1 = c_2 - c_1 \\
  y_3 &= c_3 - (y_1 + y_2) = c_3 - c_2 \\
  \end{aligned}
  $$

* Task 2: Express in matrix form:

  $$
  y = A c, \quad \text{where} \quad A = S^{-1} = \begin{pmatrix} 1 & 0 & 0 \\ -1 & 1 & 0 \\ 0 & -1 & 1 \end{pmatrix}
  $$

* Task 3: Check if columns of $S$ are independent:

  * None of the columns can be written as a linear combination of the others
  * $S$ has full rank (3)
  * The determinant of $S$ is 1, which is non-zero

  Therefore, **columns of $S$ are independent**.


**Final Answer:**

* Solutions for $y$ in terms of $c$:

  $$
  y_1 = c_1, \quad y_2 = c_2 - c_1, \quad y_3 = c_3 - c_2
  $$

* Matrix form:

  $$
  y = A c, \quad A = \begin{pmatrix} 1 & 0 & 0 \\ -1 & 1 & 0 \\ 0 & -1 & 1 \end{pmatrix}
  $$

* Column independence of $S$:

  $$
  \text{Independent}
  $$


**Topic/Idea:**

* Solving triangular linear systems
* Matrix inversion for triangular matrices
* Linear independence of columns and full rank matrices
