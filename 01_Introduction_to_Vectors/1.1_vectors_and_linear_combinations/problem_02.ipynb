### **Problem**

Let

$$
\vec{v} = \begin{bmatrix} 4 \\ 1 \end{bmatrix}, \quad \vec{w} = \begin{bmatrix} -2 \\ 2 \end{bmatrix}
$$

Draw the following vectors on the same 2D plane:

* \$\vec{v}\$
* \$\vec{w}\$
* \$\vec{v} + \vec{w}\$
* \$\vec{v} - \vec{w}\$

---

### **Solution**

#### **Step 1: Compute the vector sums and differences**

* \$\vec{v} + \vec{w} = \begin{bmatrix} 4 \ 1 \end{bmatrix} + \begin{bmatrix} -2 \ 2 \end{bmatrix} = \begin{bmatrix} 2 \ 3 \end{bmatrix}\$
* \$\vec{v} - \vec{w} = \begin{bmatrix} 4 \ 1 \end{bmatrix} - \begin{bmatrix} -2 \ 2 \end{bmatrix} = \begin{bmatrix} 6 \ -1 \end{bmatrix}\$

#### **Vectors to Plot**

* \$\vec{v} = (4, 1)\$
* \$\vec{w} = (-2, 2)\$
* \$\vec{v} + \vec{w} = (2, 3)\$
* \$\vec{v} - \vec{w} = (6, -1)\$

---
### **Code**

```python
import matplotlib.pyplot as plt
import numpy as np

# Define vectors
v = np.array([4, 1])
w = np.array([-2, 2])
v_plus_w = v + w
v_minus_w = v - w

# Create plot
plt.figure(figsize=(8, 8))
ax = plt.gca()

# Plot vectors from origin
ax.quiver(0, 0, v[0], v[1], angles='xy', scale_units='xy', scale=1, color='blue', label='v = [4, 1]')
ax.quiver(0, 0, w[0], w[1], angles='xy', scale_units='xy', scale=1, color='green', label='w = [-2, 2]')
ax.quiver(0, 0, v_plus_w[0], v_plus_w[1], angles='xy', scale_units='xy', scale=1, color='purple', label='v + w = [2, 3]')
ax.quiver(0, 0, v_minus_w[0], v_minus_w[1], angles='xy', scale_units='xy', scale=1, color='red', label='v - w = [6, -1]')

# Set plot limits
ax.set_xlim(-3, 7)
ax.set_ylim(-3, 5)

# Grid and labels
plt.grid(True)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.legend()
plt.title("Visualization of v, w, v + w, and v - w")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.gca().set_aspect('equal', adjustable='box')

plt.show()
```
