### **Problems**
* Find unit vectors in the directions of $v$ and $w$ in Problem 1, and the cosine of the angle $\theta$. Choose vectors $a$, $b$, $c$ that make $0^\circ$, $90^\circ$, and $180^\circ$ angles with $w$.

Given vectors from Problem 1:
$$
v = \begin{bmatrix} 
4 \\ 
3 
\end{bmatrix}, \quad 
w = \begin{bmatrix} 
1 \\ 
2 
\end{bmatrix}
$$

### **Solutions**

**(a) Unit vector in the direction of $\vec{v}$:**

$$
\hat{v} = \frac{\vec{v}}{|\vec{v}|} = \frac{[4, 3]}{5} = \left[\frac{4}{5}, \frac{3}{5}\right]
$$


**(b) Unit vector in the direction of $\vec{w}$:**

$$
\hat{w} = \frac{\vec{w}}{|\vec{w}|} = \frac{[1, 2]}{\sqrt{5}} = \left[\frac{1}{\sqrt{5}}, \frac{2}{\sqrt{5}}\right] \approx [0.447, 0.894]
$$


**(c) Cosine of the angle $\theta$ between $\vec{v}$ and $\vec{w}$:**

* Dot product:

  $$
  \vec{v} \cdot \vec{w} = 4 \cdot 1 + 3 \cdot 2 = 10
  $$

* Magnitudes:

  $$
  |\vec{v}| = 5, \quad |\vec{w}| = \sqrt{5}
  $$

* Cosine:

  $$
  \cos(\theta) = \frac{\vec{v} \cdot \vec{w}}{|\vec{v}| |\vec{w}|} = \frac{10}{5\sqrt{5}} = \frac{2}{\sqrt{5}} \approx 0.894
  $$

* Angle:

  $$
  \theta = \cos^{-1}(0.894) \approx 26.57^\circ
  $$


**(d) A vector $\vec{a}$ that makes a $0^\circ$ angle with $\vec{w}$:**

Any positive scalar multiple of $\vec{w}$, e.g.:

$$
\vec{a} = 2 \cdot \vec{w} = [2, 4]
$$


**(e) A vector $\vec{b}$ that makes a $90^\circ$ angle with $\vec{w}$:**

Any vector orthogonal to $\vec{w}$, e.g.:

$$
\vec{b} = [-2, 1] \quad \text{(since } \vec{w} \cdot \vec{b} = 1 \cdot (-2) + 2 \cdot 1 = 0 \text{)}
$$

**(f) A vector $\vec{c}$ that makes a $180^\circ$ angle with $\vec{w}$:**

Any negative scalar multiple of $\vec{w}$, e.g.:

$$
\vec{c} = -\vec{w} = [-1, -2]
$$
