### **Problems**
Find a basis for each of the four subspaces associated with $A$:

$$A = 
\begin{bmatrix}
0 & 1 & 2 & 3 & 4 \\
0 & 1 & 2 & 4 & 6 \\
0 & 0 & 0 & 1 & 2
\end{bmatrix} =
\begin{bmatrix}
1 & 0 & 0 \\
1 & 1 & 0 \\
0 & 1 & 1
\end{bmatrix}
\begin{bmatrix}
0 & 1 & 2 & 3 & 4 \\
0 & 0 & 0 & 1 & 2 \\
0 & 0 & 0 & 0 & 0
\end{bmatrix}$$

---

### **Solutions**

The matrix is given in its factored form $A = LU$. The second matrix is the row echelon form (U) with pivots in columns 2 and 4.

**(a) Column Space $C(A)$:**
A basis is formed by the pivot columns of the original matrix $A$.
Pivot columns: 2 and 4.
$$
\text{Basis: } \left\{ 
\begin{bmatrix}
1 \\
1 \\
0
\end{bmatrix},
\begin{bmatrix}
3 \\
4 \\
1
\end{bmatrix}
\right\}
$$

**(b) Row Space $C(A^T)$:**
A basis is formed by the non-zero rows of the row echelon form $U$.
$$
\text{Basis: } \left\{ 
\begin{bmatrix}
0 & 1 & 2 & 3 & 4
\end{bmatrix},
\begin{bmatrix}
0 & 0 & 0 & 1 & 2
\end{bmatrix}
\right\}
$$

**(c) Nullspace $N(A)$:**
Solve $A\mathbf{x} = 0$. Free variables are $x_1, x_3, x_5$.
Let $x_1=1, x_3=0, x_5=0$: $\Rightarrow \mathbf{x} = \begin{bmatrix} 1 \\ 0 \\ 0 \\ 0 \\ 0 \end{bmatrix}$.
Let $x_1=0, x_3=1, x_5=0$: From row 2: $x_2 + 2(1) = 0 \Rightarrow x_2 = -2$, row 1: $x_4 + 2(0)=0 \Rightarrow x_4=0$. $\Rightarrow \mathbf{x} = \begin{bmatrix} 0 \\ -2 \\ 1 \\ 0 \\ 0 \end{bmatrix}$.
Let $x_1=0, x_3=0, x_5=1$: From row 2: $x_4 + 2(1)=0 \Rightarrow x_4 = -2$, row 1: $x_2 + 3(-2) + 4(1)=0 \Rightarrow x_2 -6 +4=0 \Rightarrow x_2=2$. $\Rightarrow \mathbf{x} = \begin{bmatrix} 0 \\ 2 \\ 0 \\ -2 \\ 1 \end{bmatrix}$.
$$
\text{Basis: } \left\{ 
\begin{bmatrix}
1 \\
0 \\
0 \\
0 \\
0
\end{bmatrix},
\begin{bmatrix}
0 \\
-2 \\
1 \\
0 \\
0
\end{bmatrix},
\begin{bmatrix}
0 \\
2 \\
0 \\
-2 \\
1
\end{bmatrix}
\right\}
$$

**(d) Left Nullspace $N(A^T)$:**
Solve $A^T\mathbf{y} = 0$. The dimension is $m - r = 3 - 2 = 1$.
We find a vector orthogonal to the column space. Using the pivot rows from the $L$ matrix factor (which tracks the row operations), the third row of $L^{-1}$ gives the left nullspace vector.
$L = \begin{bmatrix} 1 & 0 & 0 \\ 1 & 1 & 0 \\ 0 & 1 & 1 \end{bmatrix}$, then $L^{-1} = \begin{bmatrix} 1 & 0 & 0 \\ -1 & 1 & 0 \\ 1 & -1 & 1 \end{bmatrix}$.
The last row of $L^{-1}$ is $\begin{bmatrix} 1 & -1 & 1 \end{bmatrix}$.
$$
\text{Basis: } \left\{ 
\begin{bmatrix}
1 \\
-1 \\
1
\end{bmatrix}
\right\}
$$
