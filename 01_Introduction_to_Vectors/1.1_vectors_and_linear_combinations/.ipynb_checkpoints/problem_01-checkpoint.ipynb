### **Problem**

Describe geometrically (line, plane, or all of \$\mathbb{R}^3\$) all linear combinations of:

**(a)**

$$
\begin{bmatrix} 1 \\ 2 \\ 3 \end{bmatrix} \text{ and } \begin{bmatrix} 3 \\ 6 \\ 9 \end{bmatrix}
$$

**(b)**

$$
\begin{bmatrix} 1 \\ 0 \\ 0 \end{bmatrix} \text{ and } \begin{bmatrix} 0 \\ 2 \\ 3 \end{bmatrix}
$$

**(c)**

$$
\begin{bmatrix} 2 \\ 0 \\ 0 \end{bmatrix}, \begin{bmatrix} 0 \\ 2 \\ 2 \end{bmatrix}, \text{ and } \begin{bmatrix} 2 \\ 2 \\ 3 \end{bmatrix}
$$

---

### **Solution**

**(a)**
Vectors:

$$
\vec{v}_1 = \begin{bmatrix} 1 \\ 2 \\ 3 \end{bmatrix}, \quad \vec{v}_2 = \begin{bmatrix} 3 \\ 6 \\ 9 \end{bmatrix}
$$

Notice:

$$
\vec{v}_2 = 3 \cdot \vec{v}_1
$$

So they are **linearly dependent**, and all linear combinations are scalar multiples of \$\vec{v}\_1\$.

**Geometric Description:**
The set of all linear combinations forms a **line** in \$\mathbb{R}^3\$ through the origin.


**(b)**
Vectors:

$$
\vec{v}_1 = \begin{bmatrix} 1 \\ 0 \\ 0 \end{bmatrix}, \quad \vec{v}_2 = \begin{bmatrix} 0 \\ 2 \\ 3 \end{bmatrix}
$$

These vectors are **linearly independent** (not scalar multiples of each other).
So they span a **2D subspace**.

**Geometric Description:**
The set of all linear combinations forms a **plane** in \$\mathbb{R}^3\$ through the origin.


**(c)**
Vectors:

$$
\vec{v}_1 = \begin{bmatrix} 2 \\ 0 \\ 0 \end{bmatrix}, \quad \vec{v}_2 = \begin{bmatrix} 0 \\ 2 \\ 2 \end{bmatrix}, \quad \vec{v}_3 = \begin{bmatrix} 2 \\ 2 \\ 3 \end{bmatrix}
$$

To check if they span \$\mathbb{R}^3\$, we check if they are linearly independent.
Compute the determinant of the matrix formed by these vectors as columns:

$$
\text{Let } A = 
\begin{bmatrix}
2 & 0 & 2 \\
0 & 2 & 2 \\
0 & 2 & 3
\end{bmatrix}
$$

Compute \$\det(A)\$:

$$
\det(A) = 2 \cdot \begin{vmatrix} 2 & 2 \\ 2 & 3 \end{vmatrix}
= 2(2 \cdot 3 - 2 \cdot 2) = 2(6 - 4) = 2(2) = 4 \ne 0
$$

So the vectors are **linearly independent** and span \$\mathbb{R}^3\$.

**Geometric Description:**
The set of all linear combinations forms **all of** \$\mathbb{R}^3\$.

---

### Final Answers:

* **(a)** Line
* **(b)** Plane
* **(c)** All of \$\mathbb{R}^3\$
