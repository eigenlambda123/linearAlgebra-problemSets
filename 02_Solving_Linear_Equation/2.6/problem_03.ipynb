### Problems
Forward elimination transforms $A\mathbf{x} = \mathbf{b}$ to triangular $U\mathbf{x} = \mathbf{c}$:

**Original System ($A\mathbf{x} = \mathbf{b}$):**
$$
\begin{cases}
x + y + z = 5 \\
x + 2y + 3z = 7 \\
x + 3y + 6z = 11
\end{cases}
$$

**After Elimination ($U\mathbf{x} = \mathbf{c}$):**
$$
\begin{cases}
x + y + z = 5 \\
y + 2z = 2 \\
z = 2
\end{cases}
$$

**Tasks:**
(a) Find multipliers $\ell_{31}$ and $\ell_{32}$ used in elimination.
(b) Reconstruct row 3 of $[A \ \mathbf{b}]$ using $L$ and $[U \ \mathbf{c}]$.
(c) Verify $A = LU$ and $\mathbf{b} = L\mathbf{c}$.

---

### Solutions

#### (a) Elimination Multipliers
To eliminate $x$ and $y$ in row 3:
- **$\ell_{31}$**: Eliminate $x$ (coefficient = 1) using row 1:
  $$\ell_{31} = \frac{\text{Coefficient in row 3}}{\text{Pivot in row 1}} = \frac{1}{1} = \boxed{1}$$

- **$\ell_{32}$**: Eliminate $y$ (coefficient = 2 after step 1) using row 2:
  $$\ell_{32} = \frac{2}{1} = \boxed{2}$$

#### (b) Reconstruct Row 3 of $[A \ \mathbf{b}]$
Row 3 of $A\mathbf{x} = \mathbf{b}$ is recovered by:
$$
\text{Row 3 of } [A \ \mathbf{b}] = \ell_{31} \cdot \text{Row 1} + \ell_{32} \cdot \text{Row 2} + 1 \cdot \text{Row 3}
$$
Substituting values from $[U \ \mathbf{c}]$:
$$
1 \cdot \begin{bmatrix}1 & 1 & 1 & 5\end{bmatrix} + 2 \cdot \begin{bmatrix}0 & 1 & 2 & 2\end{bmatrix} + 1 \cdot \begin{bmatrix}0 & 0 & 1 & 2\end{bmatrix} = \boxed{\begin{bmatrix}1 & 3 & 6 & 11\end{bmatrix}}
$$

#### (c) Matrix Factorization
**Lower triangular matrix $L$:**
$$
L = \begin{bmatrix}
1 & 0 & 0 \\
\ell_{21} & 1 & 0 \\
\ell_{31} & \ell_{32} & 1
\end{bmatrix}
= \boxed{\begin{bmatrix}
1 & 0 & 0 \\
1 & 1 & 0 \\
1 & 2 & 1
\end{bmatrix}}
$$

**Verification:**
- Compute $LU$:
  $$
  \begin{bmatrix}
  1 & 0 & 0 \\
  1 & 1 & 0 \\
  1 & 2 & 1
  \end{bmatrix}
  \begin{bmatrix}
  1 & 1 & 1 \\
  0 & 1 & 2 \\
  0 & 0 & 1
  \end{bmatrix}
  = \boxed{\begin{bmatrix}
  1 & 1 & 1 \\
  1 & 2 & 3 \\
  1 & 3 & 6
  \end{bmatrix}} = A \quad \checkmark
  $$
- Compute $L\mathbf{c}$:
  $$
  L\mathbf{c} = \begin{bmatrix}
  1 & 0 & 0 \\
  1 & 1 & 0 \\
  1 & 2 & 1
  \end{bmatrix}
  \begin{bmatrix}
  5 \\
  2 \\
  2
  \end{bmatrix}
  = \boxed{\begin{bmatrix}
  5 \\
  7 \\
  11
  \end{bmatrix}} = \mathbf{b} \quad \checkmark
  $$
