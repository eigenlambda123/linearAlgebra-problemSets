### **Problems**

Given the rank-one matrix $A$ and vector $b$:
$$
A = \begin{bmatrix} 
2 & 1 & 3 \\ 
6 & 3 & 9 \\ 
4 & 2 & 6 
\end{bmatrix}, \quad 
b = \begin{bmatrix} 
b_1 \\ 
b_2 \\ 
b_3 
\end{bmatrix} = \begin{bmatrix} 
10 \\ 
30 \\ 
20 
\end{bmatrix}
$$

**Tasks**:
1. Reduce $[A \mid b]$ to $[U \mid c]$ (upper triangular form)
2. Find the **two conditions** on $b_1, b_2, b_3$ for $Ax = b$ to be solvable  
   *(These conditions together put $b$ into the ______ space - hint: two planes intersect in a line)*
3. Describe the column space of $A$
4. Describe the nullspace of $A$ (special solutions in $\mathbb{R}^3$)
5. Reduce $[U \mid c]$ to $[R \mid d]$ (RREF)
6. Find the complete solution to $Ax = b$

**Note**: Matrix $A$ is constructed as an outer product:
$$
A = \begin{bmatrix} 1 \\ 3 \\ 2 \end{bmatrix} \begin{bmatrix} 2 & 1 & 3 \end{bmatrix}
$$

---
### **Solutions**

#### Step 1: Row Reduction to $\[U \mid c]\$

$$
\begin{bmatrix}
2 & 1 & 3 & | & 10 \\
6 & 3 & 9 & | & 30 \\
4 & 2 & 6 & | & 20
\end{bmatrix}
\rightarrow
\begin{bmatrix}
2 & 1 & 3 & | & 10 \\
0 & 0 & 0 & | & 0 \\
0 & 0 & 0 & | & 0
\end{bmatrix}
$$

Explanation:

* $R_2 = R_2 - 3R_1$
* $R_3 = R_3 - 2R_1$


#### Step 2: Solvability Conditions

From the reduced matrix:

$$
\text{Last two rows} = [0\ 0\ 0\ |\ 0]
$$

No new conditions are imposed — the system is **always solvable** if:

$$
\boxed{
b = \alpha \cdot \begin{bmatrix} 2 \\ 6 \\ 4 \end{bmatrix}
}
$$

Let’s check:

$$
\frac{b_2}{b_1} = \frac{30}{10} = 3, \quad \frac{b_3}{b_1} = \frac{20}{10} = 2
\quad \Rightarrow \quad
\boxed{b_2 = 3b_1,\quad b_3 = 2b_1}
$$

These are the **two conditions** needed for $b$ to lie in the column space.


#### Step 3: Column Space of $A$

Since $A$ is a rank-one matrix (outer product), all columns are in the span of $u = \begin{bmatrix} 1 \\ 3 \\ 2 \end{bmatrix}$

But each column of $A$ is also a scalar multiple of $\begin{bmatrix} 2 \\ 6 \\ 4 \end{bmatrix}$, the first column.

So:

$$
\text{col}(A) = \text{span} \left\{
\begin{bmatrix} 2 \\ 6 \\ 4 \end{bmatrix}
\right\}
$$

That’s a line in $\mathbb{R}^3$.


#### Step 4: Nullspace of $A$

We solve $Ax = 0$, with:

$$
A = 
\begin{bmatrix}
2 & 1 & 3 \\
6 & 3 & 9 \\
4 & 2 & 6
\end{bmatrix}
$$

Row reduce:

$$
R_2 = R_2 - 3R_1, \quad R_3 = R_3 - 2R_1
\Rightarrow
\begin{bmatrix}
2 & 1 & 3 \\
0 & 0 & 0 \\
0 & 0 & 0
\end{bmatrix}
$$

Divide $R_1$ by 2:

$$
R_1 = [1\ \tfrac{1}{2}\ \tfrac{3}{2}]
$$

From this:

$$
x_1 + \tfrac{1}{2}x_2 + \tfrac{3}{2}x_3 = 0 \quad \Rightarrow \quad
x_1 = -\tfrac{1}{2}x_2 - \tfrac{3}{2}x_3
$$

Let $x_2 = s$, $x_3 = t$

Then:

$$
x = \begin{bmatrix}
x_1 \\
x_2 \\
x_3
\end{bmatrix}
=
s \begin{bmatrix} -\frac{1}{2} \\ 1 \\ 0 \end{bmatrix}
+
t \begin{bmatrix} -\frac{3}{2} \\ 0 \\ 1 \end{bmatrix}
$$

So:

$$
\boxed{
\text{null}(A) = \text{span} \left\{
\begin{bmatrix} -\tfrac{1}{2} \\ 1 \\ 0 \end{bmatrix},\ 
\begin{bmatrix} -\tfrac{3}{2} \\ 0 \\ 1 \end{bmatrix}
\right\}
}
$$


#### Step 5: RREF $\[R \mid d]\$

From earlier reduction:

$$
\begin{bmatrix}
2 & 1 & 3 & | & 10 \\
0 & 0 & 0 & | & 0 \\
0 & 0 & 0 & | & 0
\end{bmatrix}
\rightarrow
\begin{bmatrix}
1 & \frac{1}{2} & \frac{3}{2} & | & 5 \\
0 & 0 & 0 & | & 0 \\
0 & 0 & 0 & | & 0
\end{bmatrix}
$$


#### Step 6: Complete Solution for $b = \begin{bmatrix} 10 \\ 30 \\ 20 \end{bmatrix}$

We already reduced to:

$$
x_1 + \frac{1}{2}x_2 + \frac{3}{2}x_3 = 5
\Rightarrow x_1 = 5 - \frac{1}{2}x_2 - \frac{3}{2}x_3
$$

Let $x_2 = s$, $x_3 = t$

##### Particular solution:

Set $s = 0,\ t = 0$ ⇒ $x_1 = 5$

So:

$$
x_p = \begin{bmatrix} 5 \\ 0 \\ 0 \end{bmatrix}
$$

##### Complete solution:

$$
x = x_p + c_1 \begin{bmatrix} -\tfrac{1}{2} \\ 1 \\ 0 \end{bmatrix} + c_2 \begin{bmatrix} -\tfrac{3}{2} \\ 0 \\ 1 \end{bmatrix}
$$
