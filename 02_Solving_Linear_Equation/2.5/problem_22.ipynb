### **Problems**
Transform $I$ into $A^{-1}$ while reducing $A$ to $I$ using row operations:

#### (a) For the first matrix:
$$
[A \mid I] = 
\left[\begin{array}{cc|cc}
1 & 3 & 1 & 0 \\
2 & 7 & 0 & 1
\end{array}\right]
$$

#### (b) For the second matrix:
$$
[A \mid I] = 
\left[\begin{array}{cc|cc}
1 & 4 & 1 & 0 \\
3 & 9 & 0 & 1
\end{array}\right]
$$

**Tasks:**  
(a) Perform row operations on the first augmented matrix to find $A^{-1}$  
(b) Perform row operations on the second augmented matrix to find $A^{-1}$

### **Solutions**

---

#### **(a) First Matrix:**
$$
[A \mid I] = 
\left[\begin{array}{cc|cc}
1 & 3 & 1 & 0 \\ 
2 & 7 & 0 & 1
\end{array}\right]
$$

**Step 1:** Eliminate the entry $a_{21} = 2$ using $R_2 \leftarrow R_2 - 2R_1$:
$$
\left[\begin{array}{cc|cc}
1 & 3 & 1 & 0 \\ 
0 & 1 & -2 & 1
\end{array}\right]
$$

**Step 2:** Eliminate the entry $a_{12} = 3$ using $R_1 \leftarrow R_1 - 3R_2$:
$$
\left[\begin{array}{cc|cc}
1 & 0 & 7 & -3 \\ 
0 & 1 & -2 & 1
\end{array}\right]
$$

**Result:**
$$
A^{-1} = \boxed{\begin{bmatrix}
7 & -3 \\
-2 & 1
\end{bmatrix}}
$$

**Verification:**
$$
AA^{-1} = \begin{bmatrix}
1 & 3 \\
2 & 7
\end{bmatrix}
\begin{bmatrix}
7 & -3 \\
-2 & 1
\end{bmatrix}
= \begin{bmatrix}
1 & 0 \\
0 & 1
\end{bmatrix} = I \quad \checkmark
$$

---

#### **(b) Second Matrix:**
$$
[A \mid I] = 
\left[\begin{array}{cc|cc}
1 & 4 & 1 & 0 \\ 
3 & 9 & 0 & 1
\end{array}\right]
$$

**Step 1:** Eliminate the entry $a_{21} = 3$ using $R_2 \leftarrow R_2 - 3R_1$:
$$
\left[\begin{array}{cc|cc}
1 & 4 & 1 & 0 \\ 
0 & -3 & -3 & 1
\end{array}\right]
$$

**Step 2:** Normalize $R_2$ by dividing by $-3$:
$$
\left[\begin{array}{cc|cc}
1 & 4 & 1 & 0 \\ 
0 & 1 & 1 & -\frac{1}{3}
\end{array}\right]
$$

**Step 3:** Eliminate the entry $a_{12} = 4$ using $R_1 \leftarrow R_1 - 4R_2$:
$$
\left[\begin{array}{cc|cc}
1 & 0 & -3 & \frac{4}{3} \\ 
0 & 1 & 1 & -\frac{1}{3}
\end{array}\right]
$$

**Result:**
$$
A^{-1} = \boxed{\begin{bmatrix}
-3 & \frac{4}{3} \\
1 & -\frac{1}{3}
\end{bmatrix}}
$$

**Verification:**
$$
AA^{-1} = \begin{bmatrix}
1 & 4 \\
3 & 9
\end{bmatrix}
\begin{bmatrix}
-3 & \frac{4}{3} \\
1 & -\frac{1}{3}
\end{bmatrix}
= \begin{bmatrix}
1 & 0 \\
0 & 1
\end{bmatrix} = I \quad \checkmark
$$
