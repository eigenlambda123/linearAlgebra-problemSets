### **Problem**  
Find the inverse $A^{-1}$ by performing row operations on the augmented matrix $[A \mid I]$ until it becomes $[I \mid A^{-1}]$:

#### **Given:**
$$
A = \begin{bmatrix}
2 & 1 & 0 \\
1 & 2 & 1 \\
0 & 1 & 2
\end{bmatrix}, \quad
[A \mid I] = 
\left[\begin{array}{ccc|ccc}
2 & 1 & 0 & 1 & 0 & 0 \\
1 & 2 & 1 & 0 & 1 & 0 \\
0 & 1 & 2 & 0 & 0 & 1
\end{array}\right]
$$

### **Solution**

#### **Step 1: Normalize the first pivot (Row 1)**
Divide $R_1$ by 2:
$$
R_1 \leftarrow \frac{1}{2} R_1
$$
$$
\left[\begin{array}{ccc|ccc}
1 & \frac{1}{2} & 0 & \frac{1}{2} & 0 & 0 \\
1 & 2 & 1 & 0 & 1 & 0 \\
0 & 1 & 2 & 0 & 0 & 1
\end{array}\right]
$$

#### **Step 2: Eliminate below the first pivot (Row 2)**
Subtract $R_1$ from $R_2$:
$$
R_2 \leftarrow R_2 - R_1
$$
$$
\left[\begin{array}{ccc|ccc}
1 & \frac{1}{2} & 0 & \frac{1}{2} & 0 & 0 \\
0 & \frac{3}{2} & 1 & -\frac{1}{2} & 1 & 0 \\
0 & 1 & 2 & 0 & 0 & 1
\end{array}\right]
$$

#### **Step 3: Normalize the second pivot (Row 2)**
Multiply $R_2$ by $\frac{2}{3}$:
$$
R_2 \leftarrow \frac{2}{3} R_2
$$
$$
\left[\begin{array}{ccc|ccc}
1 & \frac{1}{2} & 0 & \frac{1}{2} & 0 & 0 \\
0 & 1 & \frac{2}{3} & -\frac{1}{3} & \frac{2}{3} & 0 \\
0 & 1 & 2 & 0 & 0 & 1
\end{array}\right]
$$

#### **Step 4: Eliminate above and below the second pivot**
- Subtract $\frac{1}{2} \times R_2$ from $R_1$:
  $$
  R_1 \leftarrow R_1 - \frac{1}{2} R_2
  $$
- Subtract $R_2$ from $R_3$:
  $$
  R_3 \leftarrow R_3 - R_2
  $$
$$
\left[\begin{array}{ccc|ccc}
1 & 0 & -\frac{1}{3} & \frac{2}{3} & -\frac{1}{3} & 0 \\
0 & 1 & \frac{2}{3} & -\frac{1}{3} & \frac{2}{3} & 0 \\
0 & 0 & \frac{4}{3} & \frac{1}{3} & -\frac{2}{3} & 1
\end{array}\right]
$$

#### **Step 5: Normalize the third pivot (Row 3)**
Multiply $R_3$ by $\frac{3}{4}$:
$$
R_3 \leftarrow \frac{3}{4} R_3
$$
$$
\left[\begin{array}{ccc|ccc}
1 & 0 & -\frac{1}{3} & \frac{2}{3} & -\frac{1}{3} & 0 \\
0 & 1 & \frac{2}{3} & -\frac{1}{3} & \frac{2}{3} & 0 \\
0 & 0 & 1 & \frac{1}{4} & -\frac{1}{2} & \frac{3}{4}
\end{array}\right]
$$

#### **Step 6: Eliminate above the third pivot**
- Add $\frac{1}{3} \times R_3$ to $R_1$:
  $$
  R_1 \leftarrow R_1 + \frac{1}{3} R_3
  $$
- Subtract $\frac{2}{3} \times R_3$ from $R_2$:
  $$
  R_2 \leftarrow R_2 - \frac{2}{3} R_3
  $$
$$
\left[\begin{array}{ccc|ccc}
1 & 0 & 0 & \frac{3}{4} & -\frac{1}{2} & \frac{1}{4} \\
0 & 1 & 0 & -\frac{1}{2} & 1 & -\frac{1}{2} \\
0 & 0 & 1 & \frac{1}{4} & -\frac{1}{2} & \frac{3}{4}
\end{array}\right]
$$

---

### **Final Answer:**
The inverse matrix is:
$$
A^{-1} = \boxed{\begin{bmatrix}
\frac{3}{4} & -\frac{1}{2} & \frac{1}{4} \\
-\frac{1}{2} & 1 & -\frac{1}{2} \\
\frac{1}{4} & -\frac{1}{2} & \frac{3}{4}
\end{bmatrix}}
$$

**Verification:**
$$
AA^{-1} = \begin{bmatrix}
2 & 1 & 0 \\
1 & 2 & 1 \\
0 & 1 & 2
\end{bmatrix}
\begin{bmatrix}
\frac{3}{4} & -\frac{1}{2} & \frac{1}{4} \\
-\frac{1}{2} & 1 & -\frac{1}{2} \\
\frac{1}{4} & -\frac{1}{2} & \frac{3}{4}
\end{bmatrix}
= \begin{bmatrix}
1 & 0 & 0 \\
0 & 1 & 0 \\
0 & 0 & 1
\end{bmatrix} = I \quad \checkmark
$$
