### **Problem**
Find the inverses (directly or from the 2 by 2 formula) of matrices $A$, $B$, and $C$:

$$
A = \begin{bmatrix} 
0 & 3 \\ 
4 & 0 
\end{bmatrix}, \quad 
B = \begin{bmatrix} 
2 & 0 \\ 
4 & 2 
\end{bmatrix}, \quad 
C = \begin{bmatrix} 
3 & 4 \\ 
5 & 7 
\end{bmatrix}.
$$

**Tasks:**

(a) Compute the inverse of matrix $A$  
(b) Compute the inverse of matrix $B$  
(c) Compute the inverse of matrix $C$

### **Solutions**

**Given:**
The general inverse formula for a 2Ã—2 matrix $X = \begin{bmatrix} a & b \\ c & d \end{bmatrix}$ is:
$$
X^{-1} = \frac{1}{ad-bc} \begin{bmatrix} d & -b \\ -c & a \end{bmatrix} \quad \text{(when } ad-bc \neq 0\text{)}
$$

---

#### **(a) Inverse of matrix A**
$$
A = \begin{bmatrix} 
0 & 3 \\ 
4 & 0 
\end{bmatrix}
$$

1. **Compute determinant:**
   $$
   \det(A) = (0)(0) - (3)(4) = -12 \neq 0 \quad \Rightarrow \text{Inverse exists}
   $$

2. **Apply inverse formula:**
   $$
   A^{-1} = \frac{1}{-12} \begin{bmatrix} 
   0 & -3 \\ 
   -4 & 0 
   \end{bmatrix} = \begin{bmatrix} 
   0 & \frac{1}{4} \\ 
   \frac{1}{3} & 0 
   \end{bmatrix}
   $$

3. **Final answer:**
   $$
   \boxed{A^{-1} = \begin{bmatrix} 
   0 & 0.25 \\ 
   0.\overline{3} & 0 
   \end{bmatrix}}
   $$

---

#### **(b) Inverse of matrix B**
$$
B = \begin{bmatrix} 
2 & 0 \\ 
4 & 2 
\end{bmatrix}
$$

1. **Compute determinant:**
   $$
   \det(B) = (2)(2) - (0)(4) = 4 \neq 0 \quad \Rightarrow \text{Inverse exists}
   $$

2. **Apply inverse formula:**
   $$
   B^{-1} = \frac{1}{4} \begin{bmatrix} 
   2 & 0 \\ 
   -4 & 2 
   \end{bmatrix} = \begin{bmatrix} 
   \frac{1}{2} & 0 \\ 
   -1 & \frac{1}{2} 
   \end{bmatrix}
   $$

3. **Final answer:**
   $$
   \boxed{B^{-1} = \begin{bmatrix} 
   0.5 & 0 \\ 
   -1 & 0.5 
   \end{bmatrix}}
   $$

---

#### **(c) Inverse of matrix C**
$$
C = \begin{bmatrix} 
3 & 4 \\ 
5 & 7 
\end{bmatrix}
$$

1. **Compute determinant:**
   $$
   \det(C) = (3)(7) - (4)(5) = 21 - 20 = 1 \neq 0 \quad \Rightarrow \text{Inverse exists}
   $$

2. **Apply inverse formula:**
   $$
   C^{-1} = \frac{1}{1} \begin{bmatrix} 
   7 & -4 \\ 
   -5 & 3 
   \end{bmatrix} = \begin{bmatrix} 
   7 & -4 \\ 
   -5 & 3 
   \end{bmatrix}
   $$

3. **Final answer:**
   $$
   \boxed{C^{-1} = \begin{bmatrix} 
   7 & -4 \\ 
   -5 & 3 
   \end{bmatrix}}
   $$
