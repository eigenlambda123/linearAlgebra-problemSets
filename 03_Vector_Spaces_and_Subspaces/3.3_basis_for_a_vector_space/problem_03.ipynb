### **Problems**

Given the system of equations:

$$
\begin{align*}
x + 3y + 3z &= 1 \\
2x + 6y + 9z &= 5 \\
-x - 3y + 3z &= 5
\end{align*}
$$

**Task**:
Find the complete solution in the form:
$$ \mathbf{x} = \mathbf{x}_p + c\mathbf{s} $$
where:
- $\mathbf{x}_p$ is a particular solution
- $\mathbf{s}$ is a special solution in the nullspace
- $c$ is an arbitrary constant

**Instructions**:
1. Write the system in matrix form $A\mathbf{x} = \mathbf{b}$
2. Find the nullspace solution $\mathbf{s}$ to $A\mathbf{s} = \mathbf{0}$
3. Find a particular solution $\mathbf{x}_p$ to $A\mathbf{x}_p = \mathbf{b}$
4. Combine them to form the general solution

---

### **Solutions**

#### Matrix Form

We write the system:

$$
\begin{aligned}
x + 3y + 3z &= 1 \\
2x + 6y + 9z &= 5 \\
-x - 3y + 3z &= 5
\end{aligned}
$$

as:

$$
A = \begin{bmatrix}
1 & 3 & 3 \\
2 & 6 & 9 \\
-1 & -3 & 3
\end{bmatrix}, \quad
\mathbf{b} = \begin{bmatrix}
1 \\
5 \\
5
\end{bmatrix}
$$


#### Row Reduce the Augmented Matrix $\[A \mid \mathbf{b}]\$

Start:

$$
\begin{bmatrix}
1 & 3 & 3 & | & 1 \\
2 & 6 & 9 & | & 5 \\
-1 & -3 & 3 & | & 5
\end{bmatrix}
$$

##### Eliminate rows 2 and 3 using row 1:

* $R_2 \leftarrow R_2 - 2R_1$
* $R_3 \leftarrow R_3 + R_1$

$$
R_2 = [2\ 6\ 9\ |\ 5] - 2 \cdot [1\ 3\ 3\ |\ 1] = [0\ 0\ 3\ |\ 3] \\
R_3 = [-1\ -3\ 3\ |\ 5] + [1\ 3\ 3\ |\ 1] = [0\ 0\ 6\ |\ 6]
$$

So:

$$
\begin{bmatrix}
1 & 3 & 3 & | & 1 \\
0 & 0 & 3 & | & 3 \\
0 & 0 & 6 & | & 6
\end{bmatrix}
$$

##### Eliminate row 3 using row 2

* $R_3 \leftarrow R_3 - 2R_2$

$$
R_3 = [0\ 0\ 6\ |\ 6] - 2 \cdot [0\ 0\ 3\ |\ 3] = [0\ 0\ 0\ |\ 0]
$$

Final row-echelon form:

$$
\begin{bmatrix}
1 & 3 & 3 & | & 1 \\
0 & 0 & 3 & | & 3 \\
0 & 0 & 0 & | & 0
\end{bmatrix}
$$


### Back-Substitution

From second row:

$$
3z = 3 \Rightarrow z = 1
$$

From first row:

$$
x + 3y + 3z = 1 \Rightarrow x + 3y + 3 = 1 \Rightarrow x = -2 - 3y
$$

Let $y = c$ (free variable), then:

$$
x = -2 - 3c, \quad y = c, \quad z = 1
$$


### **Final Answer**

Separate the solution:

#### Particular solution: set $c = 0$

$$
\mathbf{x}_p = \begin{bmatrix} -2 \\ 0 \\ 1 \end{bmatrix}
$$

#### Special solution: subtract $\mathbf{x}_p$ from general form:

$$
\mathbf{s} = \begin{bmatrix} -3 \\ 1 \\ 0 \end{bmatrix}
$$

#### Complete solution:

$$
\boxed{
\mathbf{x} = \begin{bmatrix} -2 \\ 0 \\ 1 \end{bmatrix}
+ c \begin{bmatrix} -3 \\ 1 \\ 0 \end{bmatrix}
}
\quad \text{where } c \in \mathbb{R}
$$
