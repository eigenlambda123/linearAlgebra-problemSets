### **Problem**
For these permutation matrices, find \( P^{-1} \) by trial and error (with 1's and 0's):

#### (a) First permutation matrix:
$$
P = \begin{bmatrix}
0 & 0 & 1 \\
0 & 1 & 0 \\
1 & 0 & 0
\end{bmatrix}
$$

#### (b) Second permutation matrix:
$$
P = \begin{bmatrix}
0 & 1 & 0 \\
0 & 0 & 1 \\
1 & 0 & 0
\end{bmatrix}
$$

**Tasks:**
1. Find the inverse of the first permutation matrix by inspection
2. Find the inverse of the second permutation matrix by inspection

### **Solutions**

---

#### **(a) Inverse of the first permutation matrix**
**Given:**
$$
P = \begin{bmatrix}
0 & 0 & 1 \\
0 & 1 & 0 \\
1 & 0 & 0
\end{bmatrix}
$$

**Observations:**
1. This matrix **swaps Row 1 and Row 3** (while leaving Row 2 unchanged).
2. The inverse operation is the **same swap** (swapping them again returns the original order).

**Solution:**
The inverse is the matrix itself:
$$
P^{-1} = P = \boxed{\begin{bmatrix}
0 & 0 & 1 \\
0 & 1 & 0 \\
1 & 0 & 0
\end{bmatrix}}
$$

**Verification:**
Compute $P \cdot P^{-1}$:
$$
\begin{bmatrix}
0 & 0 & 1 \\
0 & 1 & 0 \\
1 & 0 & 0
\end{bmatrix}
\begin{bmatrix}
0 & 0 & 1 \\
0 & 1 & 0 \\
1 & 0 & 0
\end{bmatrix}
=
\begin{bmatrix}
1 & 0 & 0 \\
0 & 1 & 0 \\
0 & 0 & 1
\end{bmatrix} = I \quad \text{(Identity matrix)} \quad \checkmark
$$

---

#### **(b) Inverse of the second permutation matrix**
**Given:**
$$
P = \begin{bmatrix}
0 & 1 & 0 \\
0 & 0 & 1 \\
1 & 0 & 0
\end{bmatrix}
$$

**Observations:**
1. This matrix performs a **cyclic permutation**:
   - Row 1 → Row 2
   - Row 2 → Row 3
   - Row 3 → Row 1
2. The inverse operation is the **reverse cycle**:
   - Row 1 → Row 3
   - Row 2 → Row 1
   - Row 3 → Row 2

**Solution:**
The inverse is the transpose of $P$:
$$
P^{-1} = \boxed{\begin{bmatrix}
0 & 0 & 1 \\
1 & 0 & 0 \\
0 & 1 & 0
\end{bmatrix}}
$$

**Verification:**
Compute $P \cdot P^{-1}$:
$$
\begin{bmatrix}
0 & 1 & 0 \\
0 & 0 & 1 \\
1 & 0 & 0
\end{bmatrix}
\begin{bmatrix}
0 & 0 & 1 \\
1 & 0 & 0 \\
0 & 1 & 0
\end{bmatrix}
=
\begin{bmatrix}
1 & 0 & 0 \\
0 & 1 & 0 \\
0 & 0 & 1
\end{bmatrix} = I \quad \checkmark
$$
