### **Problems**

Given matrix $A$ and vector $b$:
$$
A = \begin{bmatrix}
2 & 4 & 6 & 4 \\
2 & 5 & 7 & 6 \\
2 & 3 & 5 & 2
\end{bmatrix}, \quad
b = \begin{bmatrix}
b_1 \\
b_2 \\
b_3
\end{bmatrix} = \begin{bmatrix}
4 \\
3 \\
5
\end{bmatrix}
$$

**Tasks**:
1. Reduce $[A \mid b]$ to $[U \mid c]$ (upper triangular form)
2. Find the condition on $b_1, b_2, b_3$ for $Ax = b$ to have a solution
3. Describe the column space of $A$ (which plane in $\mathbb{R}^3$)
4. Describe the nullspace of $A$ (special solutions in $\mathbb{R}^4$)
5. Reduce $[U \mid c]$ to $[R \mid d]$ (RREF)
6. Find complete solution to $Ax = (0, 6, -6)$

---

### **Solutions**

#### Step 1: Row Reduction to $\[U \mid c]\$

$$
\begin{bmatrix}
2 & 4 & 6 & 4 & | & b_1 \\
2 & 5 & 7 & 6 & | & b_2 \\
2 & 3 & 5 & 2 & | & b_3
\end{bmatrix}
\rightarrow
\begin{bmatrix}
2 & 4 & 6 & 4 & | & b_1 \\
0 & 1 & 1 & 2 & | & b_2 - b_1 \\
0 & -1 & -1 & -2 & | & b_3 - b_1
\end{bmatrix}
\rightarrow
\begin{bmatrix}
2 & 4 & 6 & 4 & | & b_1 \\
0 & 1 & 1 & 2 & | & b_2 - b_1 \\
0 & 0 & 0 & 0 & | & b_3 + b_2 - 2b_1
\end{bmatrix}
$$


#### Step 2: Solvability Condition

For \$Ax = b\$ to have a solution:

$$
b_3 + b_2 - 2b_1 = 0
$$


#### Step 3: Column Space

* Pivot columns: 1 and 2
* Column space is a plane in \$\mathbb{R}^3\$:

$$
\text{col}(A) = \text{span} \left\{
\begin{bmatrix} 2 \\ 2 \\ 2 \end{bmatrix},
\begin{bmatrix} 4 \\ 5 \\ 3 \end{bmatrix}
\right\}
$$


#### Step 4: Nullspace

Special solutions to \$Ax = 0\$:

1. Set \$x\_3 = 1\$, \$x\_4 = 0\$:

   $$
   \begin{bmatrix}
   1 \\ -1 \\ 1 \\ 0
   \end{bmatrix}
   $$

2. Set \$x\_3 = 0\$, \$x\_4 = 1\$:

   $$
   \begin{bmatrix}
   2 \\ -2 \\ 0 \\ 1
   \end{bmatrix}
   $$

So:

$$
\text{null}(A) = \text{span} \left\{
\begin{bmatrix} 1 \\ -1 \\ 1 \\ 0 \end{bmatrix},
\begin{bmatrix} 2 \\ -2 \\ 0 \\ 1 \end{bmatrix}
\right\}
$$


#### Step 5: RREF $\[R \mid d]\$

Reduce upper triangular to row-reduced echelon form:

$$
\begin{bmatrix}
1 & 2 & 3 & 2 & | & 0 \\
0 & 1 & 1 & 2 & | & 0 \\
0 & 0 & 0 & 0 & | & 0
\end{bmatrix}
\rightarrow
\begin{bmatrix}
1 & 0 & 1 & -2 & | & 0 \\
0 & 1 & 1 & 2 & | & 0 \\
0 & 0 & 0 & 0 & | & 0
\end{bmatrix}
$$


#### Step 6:Complete Solution for \$b = (0, 6, -6)\$

##### Step 6.1: Particular Solution

Substitute \$b = \begin{bmatrix} 0 \ 6 \ -6 \end{bmatrix}\$ into reduced system.
Solving:

$$
x_2 + x_3 + 2x_4 = 6 \Rightarrow x_2 = 6 - x_3 - 2x_4 \\
x_1 + 2x_2 + 3x_3 + 2x_4 = 0
$$

Set \$x\_3 = 0\$, \$x\_4 = 0\$:

Then \$x\_2 = 6\$,
\$x\_1 = -2(6) = -12\$

So particular solution:

$$
x_p = \begin{bmatrix} -12 \\ 6 \\ 0 \\ 0 \end{bmatrix}
$$

##### Step 6.2: Complete Solution

$$
x = x_p + c_1 \begin{bmatrix} 1 \\ -1 \\ 1 \\ 0 \end{bmatrix} + c_2 \begin{bmatrix} 2 \\ -2 \\ 0 \\ 1 \end{bmatrix}
$$
