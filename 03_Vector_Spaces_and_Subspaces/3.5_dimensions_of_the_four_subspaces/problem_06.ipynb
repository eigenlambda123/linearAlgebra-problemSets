### **Problems**

Without using elimination, find dimensions and bases for the four subspaces for  
$$A = \begin{bmatrix} 0 & 3 & 3 & 3 \\ 0 & 0 & 0 & 0 \\ 0 & 1 & 0 & 1 \end{bmatrix} \quad and  \quad B = \begin{bmatrix} 1 \\ 4 \\ 5 \end{bmatrix}.$$

### **Solutions**

**(a) Matrix A**

- Size: $3 \times 4$  

**Row space**  
- Dimension: $2$  
- Basis: $\{(0,3,3,3), (0,1,0,1)\}$  

**Column space**  
- Columns:  
  $$
  \text{col}_1 = \begin{bmatrix} 0 \\ 0 \\ 0 \end{bmatrix}, \quad
  \text{col}_2 = \begin{bmatrix} 3 \\ 0 \\ 1 \end{bmatrix}, \quad
  \text{col}_3 = \begin{bmatrix} 3 \\ 0 \\ 0 \end{bmatrix}, \quad
  \text{col}_4 = \begin{bmatrix} 3 \\ 0 \\ 1 \end{bmatrix}
  $$  

- Independent columns: $\text{col}_2, \text{col}_3$  
- Dimension: $2$  
- Basis:  
  $$
  \left\{ \begin{bmatrix} 3 \\ 0 \\ 1 \end{bmatrix}, \begin{bmatrix} 3 \\ 0 \\ 0 \end{bmatrix} \right\}
  $$  

**Nullspace**  
- Dimension: $2$  
- Basis:  
  $$
  \left\{ \begin{bmatrix} 1 \\ 0 \\ 0 \\ 0 \end{bmatrix}, \begin{bmatrix} 0 \\ 1 \\ 0 \\ -1 \end{bmatrix} \right\}
  $$  

**Left nullspace**  
- Dimension: $1$  
- Basis:  
  $$
  \left\{ \begin{bmatrix} 0 \\ 1 \\ 0 \end{bmatrix} \right\}
  $$  


**(b) Matrix B**

$$
B = \begin{bmatrix} 1 \\ 4 \\ 5 \end{bmatrix}, \quad (3 \times 1)
$$

**Row space**  
- Dimension: $1$  
- Basis: $\{(1)\}$ in $\mathbb{R}^1$  

**Column space**  
- Dimension: $1$  
- Basis:  
  $$
  \left\{ \begin{bmatrix} 1 \\ 4 \\ 5 \end{bmatrix} \right\}
  $$  

**Nullspace**  
- Dimension: $0$  
- Basis: $\{\}$  

**Left nullspace**  
- Dimension: $2$  
- Basis:  
  $$
  \left\{ \begin{bmatrix} -4 \\ 1 \\ 0 \end{bmatrix}, \begin{bmatrix} -5 \\ 0 \\ 1 \end{bmatrix} \right\}
  $$
