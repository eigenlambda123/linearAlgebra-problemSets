### **Problem**

In Problem 1, applying $E_{21}$ and then $E_{32}$ to $\vec{b} = (1, 0, 0)$ gives:

* $E_{32}E_{21}\vec{b} =$ **(a)**

Applying $E_{32}$ **before** $E_{21}$ gives:

* $E_{21}E_{32}\vec{b} =$ **(b)**

When $E_{32}$ comes first:

* Row **(c)** feels no effect from row **(d)**

---

### **Solution**

* **(a)**
  First apply $E_{21}$ to $\vec{b} = (1, 0, 0)$:

  $$
  E_{21} \vec{b} =
  \begin{bmatrix}
  1 & 0 & 0 \\
  -5 & 1 & 0 \\
  0 & 0 & 1
  \end{bmatrix}
  \begin{bmatrix}
  1 \\
  0 \\
  0
  \end{bmatrix}
  =
  \begin{bmatrix}
  1 \\
  -5 \\
  0
  \end{bmatrix}
  $$

  Then apply $E_{32}$:

  $$
  E_{32}
  \begin{bmatrix}
  1 \\
  -5 \\
  0
  \end{bmatrix}
  =
  \begin{bmatrix}
  1 & 0 & 0 \\
  0 & 1 & 0 \\
  0 & 7 & 1
  \end{bmatrix}
  \begin{bmatrix}
  1 \\
  -5 \\
  0
  \end{bmatrix}
  =
  \begin{bmatrix}
  1 \\
  -5 \\
  -35
  \end{bmatrix}
  $$

  So, **(a)** = $\begin{bmatrix} 1 \\ -5 \\ -35 \end{bmatrix}$

---

* **(b)**
  First apply $E_{32}$ to $\vec{b}$:

  $$
  E_{32} \vec{b} =
  \begin{bmatrix}
  1 & 0 & 0 \\
  0 & 1 & 0 \\
  0 & 7 & 1
  \end{bmatrix}
  \begin{bmatrix}
  1 \\
  0 \\
  0
  \end{bmatrix}
  =
  \begin{bmatrix}
  1 \\
  0 \\
  0
  \end{bmatrix}
  $$

  Then apply $E_{21}$:

  $$
  E_{21}
  \begin{bmatrix}
  1 \\
  0 \\
  0
  \end{bmatrix}
  =
  \begin{bmatrix}
  1 & 0 & 0 \\
  -5 & 1 & 0 \\
  0 & 0 & 1
  \end{bmatrix}
  \begin{bmatrix}
  1 \\
  0 \\
  0
  \end{bmatrix}
  =
  \begin{bmatrix}
  1 \\
  -5 \\
  0
  \end{bmatrix}
  $$

  So, **(b)** = $\begin{bmatrix} 1 \\ -5 \\ 0 \end{bmatrix}$

---

* **(c)** = 3 (row 3 feels no effect)
* **(d)** = 1 (because it skipped the effect of row 1)
