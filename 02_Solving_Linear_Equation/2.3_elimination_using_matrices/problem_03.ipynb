### **Problem**

Which three matrices $E_{21}, E_{31}, E_{32}$ put matrix $A$ into upper triangular form $U$?

Given:

$$
A = \begin{bmatrix}
1 & 1 & 0 \\
4 & 6 & 1 \\
-2 & 2 & 0
\end{bmatrix}
$$

We're told:

$$
E_{32} E_{31} E_{21} A = U
$$

Also, compute the matrix $M = E_{32} E_{31} E_{21}$, such that:

$$
MA = U
$$

---

### **Solution**

#### Step 1: Eliminate below pivot in column 1

* $R_2 \leftarrow R_2 - 4R_1$ → multiplier = 4
* $R_3 \leftarrow R_3 + 2R_1$ → multiplier = –2

So:

$$
E_{21} = 
\begin{bmatrix}
1 & 0 & 0 \\
-4 & 1 & 0 \\
0 & 0 & 1
\end{bmatrix}, \quad
E_{31} = 
\begin{bmatrix}
1 & 0 & 0 \\
0 & 1 & 0 \\
2 & 0 & 1
\end{bmatrix}
$$

#### Step 2: Eliminate below pivot in column 2 (after Step 1)

From matrix:

$$
\begin{bmatrix}
1 & 1 & 0 \\
0 & 2 & 1 \\
0 & 4 & 0
\end{bmatrix}
$$

We eliminate $4$ in row 3 using row 2:
$R_3 \leftarrow R_3 - 2R_2$

$$
E_{32} = 
\begin{bmatrix}
1 & 0 & 0 \\
0 & 1 & 0 \\
0 & -2 & 1
\end{bmatrix}
$$

### **Step 3: Multiply the elimination matrices**

We compute:

$$
M = E_{32} E_{31} E_{21}
$$

Start with:

$$
E_{31} E_{21} = 
\begin{bmatrix}
1 & 0 & 0 \\
0 & 1 & 0 \\
2 & 0 & 1
\end{bmatrix}
\begin{bmatrix}
1 & 0 & 0 \\
-4 & 1 & 0 \\
0 & 0 & 1
\end{bmatrix}
=
\begin{bmatrix}
1 & 0 & 0 \\
-4 & 1 & 0 \\
2 & 0 & 1
\end{bmatrix}
$$

Now:

$$
M = E_{32} (E_{31} E_{21}) = 
\begin{bmatrix}
1 & 0 & 0 \\
0 & 1 & 0 \\
0 & -2 & 1
\end{bmatrix}
\begin{bmatrix}
1 & 0 & 0 \\
-4 & 1 & 0 \\
2 & 0 & 1
\end{bmatrix}
=
\begin{bmatrix}
1 & 0 & 0 \\
-4 & 1 & 0 \\
10 & -2 & 1
\end{bmatrix}
$$

---

### Final Answer

**Elimination matrices:**

* $E_{21} = \begin{bmatrix} 1 & 0 & 0 \\ -4 & 1 & 0 \\ 0 & 0 & 1 \end{bmatrix}$
* $E_{31} = \begin{bmatrix} 1 & 0 & 0 \\ 0 & 1 & 0 \\ 2 & 0 & 1 \end{bmatrix}$
* $E_{32} = \begin{bmatrix} 1 & 0 & 0 \\ 0 & 1 & 0 \\ 0 & -2 & 1 \end{bmatrix}$

**Combined elimination matrix:**

$$
M = 
\begin{bmatrix}
1 & 0 & 0 \\
-4 & 1 & 0 \\
10 & -2 & 1
\end{bmatrix}
\quad \text{so that} \quad MA = U
$$
