### **Problems**

Find the complete solution (also called the general solution) to

$$
\begin{bmatrix}
1 & 3 & 1 & 2 \\
2 & 6 & 4 & 8 \\
0 & 0 & 2 & 4
\end{bmatrix}
\begin{bmatrix}
x \\
y \\
z \\
t
\end{bmatrix}
=
\begin{bmatrix}
1 \\
3 \\
1
\end{bmatrix}.
$$

---


### **Solution:**

We solve the system $A\mathbf{x} = \mathbf{b}$ where:
$$
A = \begin{bmatrix}
1 & 3 & 1 & 2 \\
2 & 6 & 4 & 8 \\
0 & 0 & 2 & 4
\end{bmatrix}, \quad \mathbf{b} = \begin{bmatrix} 1 \\ 3 \\ 1 \end{bmatrix}.
$$

**Row reduction to echelon form.**

Subtract 2 times row 1 from row 2:
$$
\begin{bmatrix}
1 & 3 & 1 & 2 & | & 1 \\
0 & 0 & 2 & 4 & | & 1 \\
0 & 0 & 2 & 4 & | & 1
\end{bmatrix}
$$

Now subtract row 2 from row 3:
$$
\begin{bmatrix}
1 & 3 & 1 & 2 & | & 1 \\
0 & 0 & 2 & 4 & | & 1 \\
0 & 0 & 0 & 0 & | & 0
\end{bmatrix}
$$

**Identify pivot and free variables.**
- Pivot columns: 1 and 3 (since first nonzero in row 1 is column 1, and in row 2 is column 3).
- Free variables: $y$ and $t$.

**Express pivot variables in terms of free variables.**
From row 2:
$$
2z + 4t = 1 \implies z = \frac{1 - 4t}{2}
$$

From row 1:
$$
x + 3y + z + 2t = 1
$$
Substitute $z$:
$$
x + 3y + \left(\frac{1 - 4t}{2}\right) + 2t = 1
$$
Multiply both sides by 2:
$$
2x + 6y + (1 - 4t) + 4t = 2
$$
Simplify:
$$
2x + 6y + 1 = 2 \implies 2x + 6y = 1 \implies x = \frac{1 - 6y}{2}
$$

**Write the general solution.**
Let $y = s$ and $t = u$, where $s$ and $u$ are free parameters.

Then:
$$
x = \frac{1 - 6s}{2}, \quad z = \frac{1 - 4u}{2}
$$

So the complete solution is:
$$
\begin{bmatrix} x \\ y \\ z \\ t \end{bmatrix} =
\begin{bmatrix} \frac{1}{2} \\ 0 \\ \frac{1}{2} \\ 0 \end{bmatrix} +
s \begin{bmatrix} -3 \\ 1 \\ 0 \\ 0 \end{bmatrix} +
u \begin{bmatrix} 0 \\ 0 \\ -2 \\ 1 \end{bmatrix}
$$

**Verification:**

- **Particular solution** when $s=0, u=0$: $\begin{bmatrix} \frac{1}{2} \\ 0 \\ \frac{1}{2} \\ 0 \end{bmatrix}$
- **Nullspace solutions:** $\begin{bmatrix} -3 \\ 1 \\ 0 \\ 0 \end{bmatrix}$ and $\begin{bmatrix} 0 \\ 0 \\ -2 \\ 1 \end{bmatrix}$

This matches the row-reduced system.
