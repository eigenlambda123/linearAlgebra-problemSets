### **Problems**
Solve for the first column $(x, y)$ and second column $(t, z)$ of $A^{-1}$:

$$
\begin{bmatrix}
10 & 20 \\
20 & 50
\end{bmatrix}
\begin{bmatrix}
x \\
y
\end{bmatrix}
=
\begin{bmatrix}
1 \\
0
\end{bmatrix}
\quad \text{and} \quad
\begin{bmatrix}
10 & 20 \\
20 & 50
\end{bmatrix}
\begin{bmatrix}
t \\
z
\end{bmatrix}
=
\begin{bmatrix}
0 \\
1
\end{bmatrix}.
$$

**Tasks:**  
(a) Solve for $(x, y)$ in the first system  
(b) Solve for $(t, z)$ in the second system  
(c) Combine the results to write the full inverse matrix $A^{-1}$

### **Solutions**

We find the inverse matrix $A^{-1} = \begin{bmatrix} x & t \\ y & z \end{bmatrix}$ by solving two systems of equations.

#### **Given:**
$$
A = \begin{bmatrix}
10 & 20 \\
20 & 50
\end{bmatrix}
$$

---

#### **(a) Solve for first column $\begin{bmatrix} x \\ y \end{bmatrix}$ of $A^{-1}$:**
$$
\begin{bmatrix}
10 & 20 \\
20 & 50
\end{bmatrix}
\begin{bmatrix}
x \\
y
\end{bmatrix}
=
\begin{bmatrix}
1 \\
0
\end{bmatrix}
$$

**Step-by-Step Solution:**
1. **Equation 1:** $10x + 20y = 1$  
   **Equation 2:** $20x + 50y = 0$

2. **Solve Equation 2 for $x$:**
   $$
   20x = -50y \implies x = -2.5y
   $$

3. **Substitute $x = -2.5y$ into Equation 1:**
   $$
   10(-2.5y) + 20y = 1 \implies -25y + 20y = 1 \implies y = -0.2
   $$

4. **Back-substitute to find $x$:**
   $$
   x = -2.5(-0.2) = 0.5
   $$

**Result:**
$$
\boxed{\begin{bmatrix} x \\ y \end{bmatrix} = \begin{bmatrix} 0.5 \\ -0.2 \end{bmatrix}}
$$

---

#### **(b) Solve for second column $\begin{bmatrix} t \\ z \end{bmatrix}$ of $A^{-1}$:**
$$
\begin{bmatrix}
10 & 20 \\
20 & 50
\end{bmatrix}
\begin{bmatrix}
t \\
z
\end{bmatrix}
=
\begin{bmatrix}
0 \\
1
\end{bmatrix}
$$

**Step-by-Step Solution:**
1. **Equation 1:** $10t + 20z = 0$  
   **Equation 2:** $20t + 50z = 1$

2. **Solve Equation 1 for $t$:**
   $$
   10t = -20z \implies t = -2z
   $$

3. **Substitute $t = -2z$ into Equation 2:**
   $$
   20(-2z) + 50z = 1 \implies -40z + 50z = 1 \implies z = 0.1
   $$

4. **Back-substitute to find $t$:**
   $$
   t = -2(0.1) = -0.2
   $$

**Result:**
$$
\boxed{\begin{bmatrix} t \\ z \end{bmatrix} = \begin{bmatrix} -0.2 \\ 0.1 \end{bmatrix}}
$$

---

#### **(c) Combine to form $A^{-1}$:**
$$
A^{-1} = \boxed{\begin{bmatrix}
0.5 & -0.2 \\
-0.2 & 0.1
\end{bmatrix}}
$$

**Verification:**
$$
A A^{-1} = \begin{bmatrix}
10 & 20 \\
20 & 50
\end{bmatrix}
\begin{bmatrix}
0.5 & -0.2 \\
-0.2 & 0.1
\end{bmatrix}
= \begin{bmatrix}
1 & 0 \\
0 & 1
\end{bmatrix} = I \quad \checkmark
$$
