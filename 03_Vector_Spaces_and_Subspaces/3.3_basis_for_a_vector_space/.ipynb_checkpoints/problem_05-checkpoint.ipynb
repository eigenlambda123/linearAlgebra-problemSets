### **Problems**

Under what condition on $b_1, b_2, b_3$ is this system solvable? Include $b$ as a fourth column in elimination. Find all solutions when that condition holds:

$$
\begin{aligned}
x + 2y - 2z &= b_1 \\
2x + 5y - 4z &= b_2 \\
4x + 9y - 8z &= b_3.
\end{aligned}
$$

---

### **Solutions**


We write the augmented matrix and perform Gaussian elimination:

$$
\left[\begin{array}{ccc|c}
1 & 2 & -2 & b_1 \\
2 & 5 & -4 & b_2 \\
4 & 9 & -8 & b_3
\end{array}\right]
$$

**Eliminate entries below the first pivot.**

- Subtract 2 times row 1 from row 2:
  $$
  R2 \rightarrow R2 - 2R1: \quad (2-2\cdot1, 5-2\cdot2, -4-2\cdot(-2), b_2-2b_1) = (0, 1, 0, b_2 - 2b_1)
  $$
- Subtract 4 times row 1 from row 3:
  $$
  R3 \rightarrow R3 - 4R1: \quad (4-4\cdot1, 9-4\cdot2, -8-4\cdot(-2), b_3-4b_1) = (0, 1, 0, b_3 - 4b_1)
  $$

Now the matrix is:
$$
\left[\begin{array}{ccc|c}
1 & 2 & -2 & b_1 \\
0 & 1 & 0 & b_2 - 2b_1 \\
0 & 1 & 0 & b_3 - 4b_1
\end{array}\right]
$$

**Eliminate entries below the second pivot.**

- Subtract row 2 from row 3:
  $$
  R3 \rightarrow R3 - R2: \quad (0, 0, 0, (b_3 - 4b_1) - (b_2 - 2b_1)) = (0, 0, 0, b_3 - b_2 - 2b_1)
  $$

Now the matrix is in echelon form:
$$
\left[\begin{array}{ccc|c}
1 & 2 & -2 & b_1 \\
0 & 1 & 0 & b_2 - 2b_1 \\
0 & 0 & 0 & b_3 - b_2 - 2b_1
\end{array}\right]
$$

**Solvability condition.**

For the system to be consistent, the last equation must be $0 = 0$. Therefore:
$$
b_3 - b_2 - 2b_1 = 0 \quad \text{or} \quad 2b_1 + b_2 - b_3 = 0
$$
This is the required condition on $b_1, b_2, b_3$.

**Find all solutions when the condition holds.**

Assume $2b_1 + b_2 - b_3 = 0$. Then the system becomes:
$$
\begin{aligned}
x + 2y - 2z &= b_1 \\
y &= b_2 - 2b_1
\end{aligned}
$$
Let $z$ be a free variable, say $z = t$. Then from the second equation:
$$
y = b_2 - 2b_1
$$
Substitute into the first equation:
$$
x + 2(b_2 - 2b_1) - 2t = b_1 \implies x = b_1 - 2b_2 + 4b_1 + 2t = 5b_1 - 2b_2 + 2t
$$
So the general solution is:
$$
\begin{bmatrix} x \\ y \\ z \end{bmatrix} = \begin{bmatrix} 5b_1 - 2b_2 \\ b_2 - 2b_1 \\ 0 \end{bmatrix} + t \begin{bmatrix} 2 \\ 0 \\ 1 \end{bmatrix}
$$
where $t$ is any real number.

**Verification:**
- The particular solution satisfies the system when $t=0$.
- The nullspace solution $\begin{bmatrix} 2 \\ 0 \\ 1 \end{bmatrix}$ satisfies the homogeneous system.
