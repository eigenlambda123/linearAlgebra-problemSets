### **Problem**

Given:

$$
A = \begin{bmatrix} 1 & 2 \\ 5 & 3 \end{bmatrix}, \quad 
B = \begin{bmatrix} 0 & 0 \\ 2 & 1 \end{bmatrix}, \quad 
C = \begin{bmatrix} 3 & 0 \\ 1 & 0 \end{bmatrix}
$$

Evaluate:

1. \$AB + AC\$
2. \$A(B + C)\$
   Then compare the two results.

Here is your fixed version with proper formatting and consistency throughout (correct use of `$`, `$$`, and Markdown headings):

---

## 1. \$AB + AC\$

### **Step 1: Compute \$AB\$**

$$
AB = A \cdot B = 
\begin{bmatrix} 
1 & 2 \\
5 & 3 
\end{bmatrix}
\cdot 
\begin{bmatrix}
0 & 0 \\
2 & 1 
\end{bmatrix}
=
\begin{bmatrix}
(1)(0)+(2)(2) & (1)(0)+(2)(1) \\
(5)(0)+(3)(2) & (5)(0)+(3)(1) 
\end{bmatrix}
=
\begin{bmatrix}
4 & 2 \\
6 & 3
\end{bmatrix}
$$


### **Step 2: Compute \$AC\$**

$$
AC = A \cdot C = 
\begin{bmatrix} 
1 & 2 \\
5 & 3 
\end{bmatrix}
\cdot 
\begin{bmatrix}
3 & 0 \\
1 & 0 
\end{bmatrix}
=
\begin{bmatrix}
(1)(3)+(2)(1) & (1)(0)+(2)(0) \\
(5)(3)+(3)(1) & (5)(0)+(3)(0) 
\end{bmatrix}
=
\begin{bmatrix}
5 & 0 \\
18 & 0
\end{bmatrix}
$$


### **Step 3: Compute \$AB + AC\$**

$$
AB + AC = 
\begin{bmatrix}
4 & 2 \\
6 & 3
\end{bmatrix}
+ 
\begin{bmatrix}
5 & 0 \\
18 & 0
\end{bmatrix}
=
\begin{bmatrix}
9 & 2 \\
24 & 3
\end{bmatrix}
$$

## 2. \$AB + AC\$

### **Step 4: Compute \$B + C\$**

$$
B + C = 
\begin{bmatrix}
0 & 0 \\
2 & 1 
\end{bmatrix}
+ 
\begin{bmatrix}
3 & 0 \\
1 & 0 
\end{bmatrix}
=
\begin{bmatrix}
3 & 0 \\
3 & 1
\end{bmatrix}
$$


### **Step 5: Compute \$A(B + C)\$**

$$
A(B + C) = 
\begin{bmatrix} 
1 & 2 \\
5 & 3 
\end{bmatrix}
\cdot 
\begin{bmatrix}
3 & 0 \\
3 & 1 
\end{bmatrix}
=
\begin{bmatrix}
(1)(3)+(2)(3) & (1)(0)+(2)(1) \\
(5)(3)+(3)(3) & (5)(0)+(3)(1)
\end{bmatrix}
=
\begin{bmatrix}
9 & 2 \\
24 & 3
\end{bmatrix}
$$

---

### **Final Comparison**

$$
AB + AC = \begin{bmatrix} 9 & 2 \\ 24 & 3 \end{bmatrix}, \quad
A(B + C) = \begin{bmatrix} 9 & 2 \\ 24 & 3 \end{bmatrix}
$$

✔️ They are equal:

$$
AB + AC = A(B + C)
$$

### **Conclusion: Distributive Law Holds**

$$
A(B + C) = AB + AC
$$
